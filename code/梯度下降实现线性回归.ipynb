```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1, 2, 3, 4]
y_data = [2, 4, 6, 8]

def forward(x):
    return w * x

def loss(xs, ys):
    l = 0
    for i in range(len(xs)):
        l += (ys[i] - forward(xs[i]))**2
    return l/len(xs)

def gradient(xs, ys, w):
    g = 0
    for i in range(len(xs)):
        g += 2 * (w * xs[i] - ys[i]) * xs[i]
    return g/len(xs)
```

```python
w = 1
alpha = 0.01
loss_list = []
for i in range(50):
    w -= alpha * gradient(x_data, y_data, w)
    l = loss(x_data, y_data)
    loss_list.append(l)

epoch = list(range(1, 51))
plt.plot(epoch,loss_list)
print(w)
```
