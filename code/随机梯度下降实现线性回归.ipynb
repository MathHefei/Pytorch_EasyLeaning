```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1, 2, 3, 4]
y_data = [2, 4, 6, 8]

def forward(x, w):
    return x * w

def cost(x, y, w):
    return (y - forward(x, w))**2

# 采用随机梯度下降
def gradient(x, y, w):
    g = 2 * (w * x - y)  * x 
    return g
```

```python
alpha = 0.1
w = 1
n = 40

loss_list = []
for epoch in range(n):
    for x, y in zip(x_data, y_data):
        w -= alpha * gradient(x, y, w)
        l = cost(x, y, w)
        loss_list.append(l)
        print(f"progress: {epoch}, w= {w:.2f}, loss= {l:.2f}")
epoch = list(range(1, 161))
plt.plot(epoch,loss_list)
```
