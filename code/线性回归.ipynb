```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x_data = [1, 2, 3]
y_data = [3, 5, 7]

def forward(x, w, b):
    return x * w + b

def loss(x, y, w, b):
    y_pred = forward(x, w, b)
    return (y - y_pred)**2
```

```python
w = np.arange(0,4,0.1)
b = np.arange(0,2,0.1)
w_list, b_list = np.meshgrid(w,b)

loss_list = np.zeros_like(w_list)

for i in range(len(w)):
    for j in range(len(b)):
        mse = 0
        for x, y in zip(x_data, y_data):
            mse += loss(x, y, w[i], b[j])
        loss_list[j,i] = mse/3

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(w_list, b_list, loss_list, cmap='viridis')

ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('Loss')

plt.show()
```
